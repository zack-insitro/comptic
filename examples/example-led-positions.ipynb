```python
%matplotlib notebook
%load_ext autoreload
%autoreload 2

import numpy as np
from comptic import ledarray
import matplotlib.pyplot as plt
```

## Determine which LED Arrays are Available

```python
ledarray.getAvailableLedArrays()
```

## Access LED Positions of Quasi-Dome

```python
device_name = 'quasi-dome'

# Get positions in x,y,z format
positions_cartesian = ledarray.getPositionsCart(device_name)

# Get positions in na_x, na_y format
positions_na = ledarray.getPositionsNa(device_name)

# Get board indicies
positions_boards = ledarray.getBoardIndicies(device_name)

# Get all information
positions = ledarray.getPositions(device_name)

# Show board Indicies
plt.figure(figsize=(10,4))
plt.subplot(121)
plt.title('Board Indicies')
plt.scatter(np.asarray(positions_na)[:,0], np.asarray(positions_na)[:,1], c=positions_boards)
plt.xlabel('$NA_x$')
plt.ylabel('$NA_y$')
plt.tight_layout()
plt.colorbar()

plt.subplot(122)
plt.title('LED Indicies')
plt.scatter(np.asarray(positions_na)[:,0], np.asarray(positions_na)[:,1], c=np.asarray(positions)[:,0])
plt.xlabel('$NA_x$')
plt.ylabel('$NA_y$')
plt.tight_layout()
plt.colorbar()
```
