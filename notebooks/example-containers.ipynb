# Initialize Dataset

```python
# Initialize empty Dataset
dataset = comptic.containers.Dataset()
```

## Generate Measurements

```python
# Generate random measurements
measurements = [yp.rand((100,100)) for _ in range(10)]

# Assign measurements
dataset.frame_list = measurements

# Show images
dataset.show(figsize=(5,5))
```

# Access Parameters
